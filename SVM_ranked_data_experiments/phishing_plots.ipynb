```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
import sys; sys.path.insert(0,'..')
from main import *
```

```python
X, y = load_svmlight_file('../data/phishing/phishing.txt')
X = X.toarray()

# Convert labels from {0, 1} to {-1, 1}
y = y * 2 - 1

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, stratify=y, 
                                                    shuffle=True, random_state=1)
```

## SD

```python
svm_arr, tree_arr = data_size_experiment(X_train, y_train, X_test, y_test, 
                                         experiment_type='SD', kernel='linear', 
                                         random_state=1, verbose=True)
```

```python
np.save('temp_arrays/SD/phishing_svm', svm_arr)
plot_err(svm_arr, 'Phishing', 'SVM', experiment_type='SD', scale=1, save=True)
```
