```python
X_train, y_train = load_svmlight_file('../data/adult/a5a.txt', n_features=123)
X_train = X_train.toarray()

X_test, y_test = load_svmlight_file('../data/adult/a5a.t')
X_test = X_test.toarray()
```

```python
svm_arr, tree_arr = data_size_experiment(X_train, y_train, X_test, y_test,
                                         rank=True, rounded=True, plots_or_table='labeled',
                                         random_state=2, verbose=True)
```

```python
from matplotlib import pyplot as plt

plt.style.use('seaborn-whitegrid')
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['figure.figsize'] = [5, 3]

arr = svm_arr
scale = 1

fig = plt.figure()
ax = fig.add_subplot()

NUM_REPEATS = 10

x = np.array([50, 100, 200, 400, 600, 800, 1600])
y = np.mean(arr, axis=1)

sderr = np.std(arr, axis=1)
sderr = sderr / np.sqrt(NUM_REPEATS)

ax.plot(x, y[:, 1], c='C1')
ax.fill_between(x, y[:, 1] - (sderr[:, 1] * scale), y[:, 1] + (sderr[:, 1] * scale),
             facecolors='C1', alpha=0.5)
ax.scatter(x, y[:, 1], label='PairBoost', c='C1')

ax.plot([x[0], x[-1]], [1 - 0.844, 1 - 0.844], c='black', linestyle='--', label='Supervised Basline')

ax.set_xlabel('m')
ax.set_ylabel('Classification Error')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], frameon=True ,shadow=True, fancybox=True, framealpha=1.0)

fig.savefig('final_plots/m/adult.pdf', bbox_inches='tight')
```
